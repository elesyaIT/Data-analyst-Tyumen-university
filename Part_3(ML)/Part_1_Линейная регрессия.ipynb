```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/content/drive/MyDrive/Tyumen_University/data/students.csv")
```

```python
df.info()
```

```python
df_cut = df[["Growth","Shoe size"]]
df_cut = df_cut.dropna()
```

```python
sns.scatterplot(data=df_cut, x="Shoe size", y="Growth")
```

```python
lr = LinearRegression()
results = lr.fit(df_cut["Shoe size"].values.reshape(-1,1), y=df["Growth"].values)
```

```python
results.coef_, results.intercept_
```

```python
df_cut
```

```python
df_cut["Predicted Growth"]= results.predict(df_cut["Shoe size"].values.reshape(-1,1))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(df_cut["Growth"], df_cut["Predicted Growth"])
```

```python
df_cut.head(20)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(df_cut["Growth"], df_cut["Predicted Growth"])
```
